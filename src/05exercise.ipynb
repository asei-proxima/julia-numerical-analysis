```julia
using Plots
# 問題に含まれる定数
γ = 1.0

Up(u) = (π - u )/ ( γ * sin(u)^(1.2) )

plot( 0:0.01:π, Up ) # u_n の範囲は 0 より大きく π より小さいので．
```

```julia
DUp(u) = - (sin(u)^1.2 + 1.2 * (π-u)* sin(u)^0.2 * cos(u)) / (sin(u)^1.2)
plot( 0:0.01:π, DUp )
```

```julia
using NLsolve

function nls(func, params...; ini = [0.0])
    if typeof(ini) <: Number
        r = nlsolve((vout,vin)->vout[1]=func(vin[1],params...), [ini])
        v = r.zero[1]
    else
        r = nlsolve((vout,vin)->vout .= func(vin,params...), ini)
        v = r.zero
    end
    return v, r.f_converged
end
```

```julia
f(x) = cos(x) - 1/7
sol = nls(f, ini = 1.0)  # ini = 1.0 は近似根 x の探索初期値．
```

```julia
f(u) = - (sin(u)^1.2 + 1.2 * (π-u)* sin(u)^0.2 * cos(u)) / (sin(u)^1.2)
sol = nls(f, ini = 1.0)
```
